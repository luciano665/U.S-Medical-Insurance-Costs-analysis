```python
#Function to open files and store in list created
def read_info(list_to_add, type_info):
    list_pre = []
    #open method use, read icsv file
    with open("insurance.csv") as insurance_file:

        insurance_read = csv.DictReader(insurance_file)

        for row in insurance_read:
            list_pre.append(row[type_info])
            list_to_add = list_pre

        return list_to_add
```

```python
#Call functio and store info to new lists
ages_final = read_info(ages, "age")
print(ages_final)
sex_final = read_info(sex, "sex")
print(sex_final)
bmis_final = read_info(bmis, "bmi")
print(bmis_final)
childrens_final = read_info(childrens, "children")
print(childrens_final)
smoker_final = read_info(smoker, "smoker")
print(smoker_final)
regions_final = read_info(regions, "region")
print(regions_final)
costs_final = read_info(prices, "charges")
print(costs_final)
```

```python
#Average func for data expressed as numbers
def average_list_float(arr):
    total_sum = 0
    if len(arr) == 0:
        return print("No data avialable to use " + str(arr))
    for i in range(len(arr)):
        total_sum += float(arr[i])
    
    total_average = total_sum / len(arr)
    return total_average

average_ages_final = round(average_list_float(ages_final), 3)
average_insurance_cost = (average_list_float(costs_final), 3)
average_bmi = (average_list_float(bmis_final), 3)
average_num_children = round(average_list_float(childrens_final), 3)
#print(average_num_children)
#print(average_insurance_cost)
#print(average_ages_num)
#print(average_bmi)

```

```python
#function for regions to get average for gender smoker and nono smoker
def for_smo_and_g(dict, indx ,k1, k2):
    total_males = 0
    total_females = 0
    total_smokers = 0
    total_nonSmokers = 0
    dict_gender = {}
    dict_smoke = {}
    for i in dict.values():
        if i[indx] == k1 and indx == 1:
            total_females += 1
        elif i[indx] == k2 and indx == 1:
            total_males += 1
        elif i[indx] == k1 and indx == 4:
            total_smokers +=1
        elif i[indx] == k2 and indx == 4:
            total_nonSmokers += 1
    
    if indx == 1:
        dict_gender["males"] = total_males
        #dict_gender["males_ave"] = round(total_males / len(dict))
        dict_gender["females"] = total_females
        #dict_gender["females_ave"] = round(total_females / len(dict))
        return dict_gender
    elif indx == 4:
        dict_smoke["Smokers"] = total_smokers
        #dict_smoke["Smoke_ave"] = round(total_smokers / len(dict))
        dict_smoke["Non_smokers"] = total_nonSmokers
        #dict_smoke["Non_smokers_ave"] = round(total_nonSmokers / len(dict))
        return dict_smoke
    
```

# Segmentation for insurance costs
    #Segmentation base on smoker, non-smoker, with children and without childrens
    #Get average of diffrent scenarions:
        #Insurace cost of smokers and non-smokers only
        #Insurance cost if ther eis childrens and if not
        #Insurance cost of combination of smoker/non-smoter with childrens and without childrens
    #This is segemented for each region, to compare correlation of costs but not causation

```python
#Function to check de insurance cost for smokers/non-smokers and for people with childrens and not
def average_cost_for_diff_categ(dict, idx, k1_smoke, k2):
    list_smok = []
    list_nonS = []
    list_childrens = []
    list_not_childr = []
    list_smoke_withchil = []
    list_smoker_notchil = []
    list_nonS_withchill = []
    list_nonS_nochill = []

    for i in dict.values():
        if i[k1_smoke] == "yes":
            if idx == -1:
                list_smok.append(i[-1])
            elif i[idx] != '0' and idx == 3:
                list_smoke_withchil.append(i[-1])
            elif i[idx] == '0' and idx == 3:
                list_smoker_notchil.append(i[-1])
                
        elif i[k1_smoke] == "no":
            if idx == -1:
                list_nonS.append(i[-1])
            elif i[idx] != '0' and idx == 3:
                list_nonS_withchill.append(i[-1])
            elif i[idx] == '0' and idx == 3:
                list_nonS_nochill.append(i[-1])
        
        if idx == 3:
            if i[idx] != '0':
                list_childrens.append(i[-1])
            else:
                list_not_childr.append(i[-1])


    

    #For returning lists
    if k2 == 11:
        return list_smok
    elif k2 == 12:
        return list_nonS
    elif k2 == 13:
        return list_childrens
    elif k2 == 14:
        return list_not_childr
    elif k2 == 15:
        return list_nonS_withchill
    elif k2 == 16:
        return list_nonS_nochill
    elif k2 == 17:
        return list_smoke_withchil
    elif k2 == 18:
        return list_smoker_notchil
    



```
